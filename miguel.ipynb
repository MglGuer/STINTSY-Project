```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('LFS PUF April 2016.csv')  

print("Dataset preview:")
print(df.head())
```

```python
print("Dataset dimensions:", df.shape)
print("Data types:\n", df.dtypes)
```

Data Cleaning

```python
columns_with_whitespace = df.apply(lambda col: col.str.isspace().sum() if col.dtype == 'object' else 0)

print("Number Empty Cells:")
print(columns_with_whitespace[columns_with_whitespace > 0])
```

Distribution by Region

```python
import seaborn as sns

sns.countplot(data=df, x='PUFREG', palette='viridis', order=df['PUFREG'].value_counts().index)
plt.title('Distribution of PUFREG')
plt.xlabel('Regions')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

Age and Sex Dist

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sex_counts = df['PUFC04_SEX'].value_counts()
sns.countplot(data=df, x='PUFC04_SEX', palette='coolwarm', order=sex_counts.index, ax=axes[0])

axes[0].set_title('Distribution of Sex (PUFC04_SEX)')
axes[0].set_xlabel('Sex')
axes[0].set_ylabel('Count')

axes[1].hist(df['PUFC05_AGE'], bins=20, color='lightblue', edgecolor='black')
axes[1].set_title('Age Distribution (PUFC05_AGE)')
axes[1].set_xlabel('Age')
axes[1].set_ylabel('Frequency')

plt.tight_layout()

plt.show()
```

Working hours

```python

sns.histplot(df['PUFC18_PNWHRS'], kde=True, bins=20, color='skyblue')
plt.title('Distribution of Working Hours (PUFC18_PNWHRS)')
plt.xlabel('Working Hours')
plt.ylabel('Frequency')
plt.show()
```
